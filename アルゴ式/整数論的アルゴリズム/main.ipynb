### 素数判定

```python
def is_Prime(N):
    if N == 1:
        return False
    for i in range(2, int(N ** (1 / 2)) + 1):
        if N % i == 0:
            return False
    return True

N = int(input())
print('Yes' if is_Prime(N) else 'No')
```

### エラトステネスの篩

```python
def Eratosthenes(N):
    isprime = [True] * (N + 1)
    isprime[0] = False
    isprime[1] = False

    for p in range(2, N + 1):
        if not isprime[p]:
            continue
        
        q = p * 2
        while q <= N:
            isprime[q] = False
            q += p

    return isprime
    

N = int(input())
isprime = Eratosthenes(N)
print(isprime.count(True))
```

### 約数

```python
def calc_devisors(N):
    res = []
    for i in range(1, N + 1):
        if i * i > N:
            break
        if N % i != 0:
            continue
        res.append(i)
        if N // i != i:
            res.append(N // i)

    res.sort()
    return res

N = int(input())
res = calc_devisors(N)
print(len(res))
```

### 素因数分解

```python
def prime_function(N):
    res = []
    for p in range(2, N):
        if p * p > N:
            break
        if N % p != 0:
            continue

        e = 0
        while N % p == 0:
            e += 1
            N //= p

        res.append((p, e))
        
    if N != 1:
        res.append((N, 1))
    return res

N = int(input())
res = prime_function(N)
# print(res)

for tmp in res:
    for i in range(tmp[1]):
        print(tmp[0], end=' ')
print()
        
```
